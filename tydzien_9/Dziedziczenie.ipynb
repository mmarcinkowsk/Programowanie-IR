# Dziedziczenie

Dziedziczenie polega na tworzeniu nowych klas w oparciu o już istniejące. Klasa dziedzicząca ma wtedy dostęp do wszystkich pól i metod klasy matki, przy czym może je nadpisywać. Po nadpisaniu danej metody, można nadal odwołać się do jej oryginalnej przez `super()`.

```python
class Zwierze():

    def __init__(self):
        print('Tworzę obiekt klasy Zwierze')

class Ssak(Zwierze): # klasa Ssak dziedziczy po klasie Zwierze

    def __init__(self, imie):
        super().__init__() # wywołuję konstruktor klasy Zwierze w trakacie tworzenia obiektu klasy Ssak
        self.imie = imie
        print(f'{self.imie} jest ssakiem.')

    def czy_cieplokrwisty(self):
        return True

class Pies(Ssak):
    
    def __init__(self, imie, rasa):
        super().__init__(imie) # wywołuję konstruktor klasy Ssak w trakcie tworzenia obiektu klasy Pies.and
        # Ponieważ konstruktor Klasy Ssak przyjmuje argument imie, muszę go podać temu kosntruktorowi.
        self.rasa = rasa


Fafik = Pies('Fafik', 'Dalmatyńczyk') # konstruktor wywołuje wszystkie wcześniejsze konstruktory z powodu wykorzystania super()
print(Fafik.imie) # Ponieważ Pies dziedziczy po Ssak, Fafik posiada pole .imie, odziedziczone po Ssak
print(Fafik.rasa)
print(Fafik.czy_cieplokrwisty()) # Podobnie Fafik odziedziczł metodę .czy_cieplokrwisty, którtej nie nadpisał, więc robi to samo, co w klasie Ssak
```

Fukncja `isinstance(obiekt, Klasa)` sprawdza nie tylko, czy `obiekt` należy do klasy `Klasa`, ale również, czy do klasy dziedziczącej po `Klasa`.

```python
print(isinstance(Fafik, Pies))
print(isinstance(Fafik, Ssak)) # Pies dziedziczy po Ssak, więc to również zwróci prawdę
```

# Rozpakowywanie listy/krotki jako argumentów funkcji

Jeśli mamy funkcję `f(x1, x2, ..., xn)` przyjmującą $n$ argumentów oraz listę/krotkę `x` zawierającą $n$ elementów i chcemy wywołać tę funkcję na tych elementach, możemy rozpakować ją do tych argumentów notacją z gwiazdką: `f(*x)`.

```python
def f(x, y, z):
    print(x, y, z)

a = [1, 2, 3]

f(*a)
```

Podobnie możemy zapełnić tylko część argumentów.

```python
def g(x, y, z, w):
    print(x, y, z, w)

a = (1, 2)

g(3, 4, *a)
g('Ala', *a, 'ma kota')
g(*a, 7, 8)
```

# Odowłania do list (i innych obiektów zmienniczych)
Jeśli `a` to lista, to pisząc `b=a` nie tworzymy nowej listy `b` zawierającej kopię elementów `a`, tylko nazwą `b` odwołujemy się do **tej samej listy**. W ten sposób możemy na przykład zrobić funkcję, która przyjmuje jakąś listę, i modyfikuje jej zawartość, bez potrzeby zwracania nowego obiektu. Nie marnujemy też pamięci i mocy obliczeniowej komputera na tworzenie wielu kopii tego samego obiektu.

W zrozumieniu C/C++ odpowiada to odwoływaniu się do listy przez wskaźnik/referencję, a nie przez wartość.

Niestety efektem ubocznym tego zachowania są niespodziewane i trudne do wyłapania błędy w logice programu.

```python
a = [1, 2]
b = a
b[1], b[0] = b[0], b[1] # zmianiam kolejność elementów b
print(b)
print(a) # ponieważ a i b odwołują się do tego samego obiektu, tu też widzimy zamienoiną kolejność
```

Można to obejść na kilka sposobów. Na przykład skopiować całą zawartość listy:

```python
a = [1, 2]
b = a[:] # kopiuję wycinek a zawierający wszyskie elementy

b[1], b[0] = b[0], b[1]
print(b)
print(a)
```

...albo explicite stworzyć nową listę funkcją `list`:

```python
a = [1, 2]
b = list(a) # konstruktory jak list zawsze tworzą nowy obiekt

b[1], b[0] = b[0], b[1]
print(b)
print(a)
```

...albo użyć niezwmienniczych obiektów, jak krotki (nie szkodzi, że odwołujemy się do tego samego obiektu, bo i tak nie możemy go zmodyfikować).
