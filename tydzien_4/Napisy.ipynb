# Łańcuchy tekstowe (stringi)
W Pythonie napisy otacza się pojedynczymi lub podwójnymi cudzysłowami. (Wiele innych języków i pliki `json` uznają tylko podwójne cudzysłowy.)

```python
print("Hello world!")
print('Hello world!')
```

Napisy można ze sobą sklejać operacją dodawania.

```python
print("Ala " + "ma kota.")
```

Można też je powielać, mnożąc je przez liczbę całkowitą.

```python
print('abc'*5)
```

Napisy są konenerami znaków (liter) i podobnie jak listy i krotki są indeksowane liczbami.

```python
a = 'Ala ma kota.'
print(a[1])
print(a[4:9])
```

`len()` w działaniu na napisie zwraca liczbę znaków.

```python
print(len('Ala ma kota.'))
```

### Znaki modyfikacji (_escape charaters_)
W napisach można stosować szereg specjalnych znaków, zazwyczaj zaczynających się od ukośnika `\`, które pełnią specjalne funkcje, na przykład:
- `\n` - znak nowej linii,
- `\t` - znak tabulatora,
- `\"` - cudzysłów wewnątrz napisu (nie kolidujący z cudzysłowami zamykającymi napis),
- `\\` - ukośnik nie zaczynający znaku specjalnego.

Pełna lista znajduje się w dokumentacji: https://docs.python.org/3/reference/lexical_analysis.html#escape-sequences.

```python
print("Ala\nma\t\"kota\".\\")
```

Jeśli zamiast tego chcemy, by znaki modyfikacji nie były interpretowane, możemy przed otwarciem łańcucha dopisać przedrostek `r` jak _raw_ (surowy).

```python
print(r"Ala\nma\t\"kota\".\\")
```

Przedrostek `r` można łączyć z przedroskiem `f` od formatowania.

```python
print(rf"Ala\nma\t {2+3}\"kotów\".\\")
```

Jeśli w napisach z przedroskiem `f` chcemy użyć nawiasów klamrowych `{}` nie oznaczających formatownia, musimy pisać je podwójnie `{{` i `}}` (jak `\\`).

```python
print(f"Ala\nma\t {{2+3}}\"kotów\".\\")
```

**Uwaga** Większość kombinacji `\znak` nie ma specjalnej funkcji. W wersji Pythona zainstalowanej na komputerach OKWF (w 2026 roku to Python 3.10) są pozostawiane jako znak `\` i niezmieniony znak `znak`. Od Pythona 3.12 nieropoznane kombinacje `\znak` wywołują ostrzerzenie `SyntaxWarning`, ale nadal są interpretowane tak samo. W przyszłej wersji Pythona takie napisy będą powodować błąd. Dlatego najlepiej zapisywać wszystkie znaki `\` nie mające dokonać modyfikacji jako `\\`, lub stosować przedrostek `r`.

To jest szczególnie ważne dla osób piszących programy pod Windowsa, bo w nim znaku `\` używa się do rozdzielania katalogów przy precyzowaniu ścieżki do pliku.

### Napisy na wiele linijek

Lańcuch tekstowy możemy otoczyć trzema cydzysłowami (pojedynczymi lub podwójnymi). Wtedy możemy łamać normalnie linie, zmamiast używania `\n`. 

```python
print('''Litwo! Ojczyzno moja! ty jesteś jak zdrowie:
Ile cię trzeba cenić, ten tylko się dowie,
Kto cię stracił.''')
```

Napisy wielolinijkowe normalnie wspierają przedroski `r` i `f`.

# Metody łańcuchów

Łańuchy tekstowe posiadają szereg metod, opisanych w dokumentacji https://docs.python.org/3/library/stdtypes.html#text-and-binary-sequence-type-methods-summary.

Wywołuje się je składnią `zmienna_tekstowa.metoda(argumenty)` i zwracają one wynik (nie zmieniają samej zmiennej `zmienna_tekstowa`).

Kilka przykładów:

```python
print('Ala ma kota.'.split('a')) # pocięcie napisu po konkretnym znaku
napis = '1,2,3,4,5,a,b,c,defg,h'
print(napis.split(','))
```

```python
print('Ala ma fokę.'.upper()) # zmiana wszystkich liter na wielkie (obsługuje polskie znaki)
```

```python
print('Ala ma kota.'.rjust(20)) # wyjustowanie napisu do prawej, by miał określoną szerokość
print('^'*20)
```

```python
print('Ala ma kota.'.isnumeric()) # sprawdzenie, czy napis jest liczbą
```

```python
print('rysunki/wykres.pdf'.endswith('.pdf')) # sprawdzenie, czy końcówka napisu to inny napis
```

```python
print('Ala ma kota.'.replace('kota', 'psa')) # podmiana fragmentu napisu
```

```python
print('!!!!!!!!Ala ma kota.___111____'.strip('_!1')) # ucięcie niechcianych znaków z końców napisu
```

```python
print('Ala ma kota.'.index('ma')) # znalezienie położenia fragmentu napisu wewnątrz napisu
```

Wiele z tych metod posiada opcjonalne argumenty modyfikujące ich działanie. Na przykład metoda `str.replace(old, new)` normalnie podmienia wszytkie napisy `old` na napisy `new`, ale jako trzeci argument można jej podać liczbę mówiącą, po ilu podmianach ma przestać.

Metod jest zbyt wiele, by wszyskie zapamiętać, razem z opcjonalnymi argumentami, dlatego polecam zapamiątać tylko mniej więcej jakich metod można się spodziewać i nauczyć się czytać dokumentację.

Listę metod można też wyświetlić w samym Pythonie, wykorzystując funkcję `dir`:

```python
print(dir(str))
```

A ich zastosowanie sprawdzić funkcją `help`:

```python
help(str.ljust)
```
