```python
class Klasa():
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.xkwadrat = x**2
        
    def __str__(self):
        return f'x={self.x}, y={self.y}'
        
    def __add__(self, other):
        if isinstance(other, Klasa):
            return Klasa(self.x + other.x, self.y + other.y)
        elif isinstance(other, (int, float)):
            return Klasa(self.x + other, self.y + other)
        else:
            raise TypeError()
        
    def __radd__(self, other):
        return self + other
        
    def __sub__(self, other):
        if isinstance(other, Klasa):
            return Klasa(self.x - other.x, self.y - other.y)
        
    def __iadd__(self, other):
        if isinstance(other, Klasa):
            return Klasa(self.x + other.x, self.y + other.y)
        
    def __eq__(self, value):
        if isinstance(value, Klasa):
            return self.x == value.x and self.y == value.y
        else:
            return False
        
    def __ge__(self, value):
        if isinstance(value, Klasa):
            return self.x >= value.x
        
    def __getitem__(self, key):
        if key == 'x':
            return self.x
        elif key == 'y':
            return self.y
        else:
            raise KeyError()
        
    def __setitem__(self, key, value):
        if key == 'x':
            self.x = value
        elif key == 'y':
            self.y = value
        else:
            raise KeyError()
        
    def __call__(self, *args):
        print(f'Obiekt {self} został wywołany z argumentami {args}')
        
    def __float__(self):
        return float(self.x + self.y)

        
a = Klasa(2, 5)
b = Klasa(3, -12)
c = a + b
# print(c.x, c.y, c.xkwadrat)
print(c)
print(a - b)
a += b
print(a)
print(b + 128.3)
print(128.3 + b)
# print(b + 'abc')
print(a)
a['x'] = 99
print(a)
a(123)
float(a)
```

```python
import numpy as np

class FiguraGeometryczna():
    
    def __init__(self):
        pass
    
    def pole(self):
        raise NotImplementedError()

class Elipsa(FiguraGeometryczna):
    
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    def pole(self):
        return self.a*self.b*np.pi
    
class Okrag(Elipsa):
    
    def __init__(self, r):
        self.r = r
        super().__init__(r, r)
    
    
e = Elipsa(3, 4)
e.pole()

o = Okrag(1)
o.pole()
```
