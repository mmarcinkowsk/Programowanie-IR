# Pliki (głównie tekstowe)

W Pythonie pliki otwiera się za pomocą funkcji `open(scierzka_do_pliku)`, która zwraca obiekt reprezentujący plik.

Po zakończeniu pracy zamyka się plik wykonując na tym obiekcie metodę `.close()`. W przypadku nadpisywania plików, zapisanie pliku nastąpi w tym momencie.

```python
plik = open('inwokacja.txt')

print(plik.read())

plik.close()
```

Funkcja `open()` jako drugi (opcjonalny) argument przyjmuje tryb otwarcia pliku. Standardową wartością jest `'r'`, czy _tylko do odczytu_. Można podać tryb `'w'`, by otworzyć plik w trybie _tylko do nadpisywania_.

Tryb `'w'` nadpisuje istniejący plik. Jeśli otworzymy w trybie `'w'` plik, który nie istnieje, zostanie on utworzony.

Istnieje też tryb `'a'`, w którym dopisujemy nową treść na końcu istniejącego pliku.

```python
plik = open('out.txt', 'w')

plik.write("Ala ma kota.")

plik.close()
```

Nie można otworzyć pliku jednocześnie w trybie `'r'` i `'w'`. Zamiast tego, należy otworzyć go w trybie odczytu, zapamiętać jego treść w zmiennej/-ych, zamknąć plik, dokonać odpowiednich przekształceń i dopiero wtedy otworzyć go ponownie w tybie zapisu i nadpisać. Dopiero, gdy zamkniemy go ponownie, zostanie on zapisany na dysku.

Istnieje też alternatywna składnia do otwierania plików, wykorzysująca wcięcia, jak wyrażenie `if` czy pętle:

```python
with open('inwokacja.txt') as plik:
    print(plik.read())
    
print('Plik zamknięty')
```

Korzystając z konstrukcji `with`, plik jest dostępny we wciętym bloku kodu, a gdy wcięcie się kończy, jest automatycznie zamykany.

Oba sposoby są równie dobre, to, z którego korzystamy zależy od własnych preferencji. Czasem też specyfika rozwiązywanego problemu czyni jeden z tych sposobów wygodniejszym.

### Metody plików tekstowych

Najczęściej stosowane metody plików tekstowych:

W trybie odczytu (`'r'`):
- `plik.read()` - zwraca całą treść pliku jako jeden wielki napis,
- `plik.readlines()` - zwraca listę zawierającą kolejne wiersze odczytane z pliku,
- `plik.readline()` - przy pierwszym wywołaniu zwraca pierwszy wiersz pliku. Przy drugim zwraca drugi wiersz itd.

W trybie zapisu (`'w'`):
- `plik.write(tekst)` - dopisuje `tekst` do pliku. Można wywoływać tę metodę wielokrotnie, za każdym razem dopisując nowy kawałek tekstu.

```python
with open('inwokacja.txt') as plik:
    i = 0
    for linia in plik.readlines():
        print(i, linia)
        i += 1
```

### `__file__`
Skrypty w Pythonie posiadają predefiniowaną stałą `__file__`, która zawiera ścierzkę do pliku z kodem źródłowym. Stała `__file__` nie jest dostępna w notatnikach jupytera, dlatego nie mogę jej tu pokazać. Jej przydatne zastosowanie pokazuję w pliku [uzycie__file__.py](./uzycie__file__.py).
