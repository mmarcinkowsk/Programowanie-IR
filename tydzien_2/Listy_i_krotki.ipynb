# Listy i krotki

Listy i krotki są kontenerami - służą do przechowywania innych obiektów.

### Listy
Listy tworzy się, zamykając obiekty oddzielone przecinkami w kwadratowe nawiasy. Listy mogą zawierać dowolne obiekty i zapamiętują ich kolejność.

````python
lista = [6, 27, -9, 14, 33, 28, 1, 1, "Ala ma kota", [1, 2, 3], None, print, 6]
print(lista)
````

Do elementów listy można odwołać się po ich indeksie (numerze). Elementy listy numerowane są od zera.

````python
print(lista[2])
lista[1] = 8
print(lista[1])
````

Dodatkowo python pozwala używać ujemnych indeksów, by odwoływać się do elementów listy od końca. Wtedy -1 to ostani, -2 przedostatni itp.

````python
print(lista[-1])
print(lista[-2])
````

Można też odwoływać się do fragmentu listy ograniczonego indeksami.

````python
print(lista[2:6])
````

Funkcja `len()` wywołana na liście zwraca jej długość.

````python
print(len(lista))
````

Metoda `.append()` doczepia nowy element na końcu listy.

````python
lista.append(19)
print(lista)
````

Metoda `.pop()` usuwa element listy po jego indeksie. Przy okazji go zwraca.

````python
ósmy_element = lista.pop(8)
print(lista)
print(ósmy_element)
````

`.pop()` wywołane bez sprecyzowania indeksu zwraca ostatni element.

````python
print(lista.pop())
print(lista)
````

Metoda `.remove()` usuwa element po jego wartości.

````python
lista.remove(14)
print(lista)
````

Gdy dany element się powtarza, `.remove()` usuwa tylko pierwszy.

````python
lista.remove(1)
print(lista)
````

### Pętla `for` po elementach listy
Po elementach listy można zrobić pętlę `for`.

````python
for x in lista:
    print(x)
````

### Wyrażenia listowe
Bardzo wygodnym sposobem tworzenia list są tak zwane wyrażenia listowe. Polegają one na "zamknięciu" w listę pętli `for`, która ją wypełnia.

````python
parzyste = [2*n for n in range(10)]
print(parzyste)
````

### Krotki
Krotki są podobne do list, z tą różnicą, że utworzona krotka nie może ulec zmianie. Dlatego też krotki nie posiadają metod `.append()`, `.pop()` i `.remove()`.
Krotki otacza się okrógłymi nawiasami zamiast kwadratowych.

````python
krotka = (1, 2, 8, 19, -6)
print(krotka[2])
````

Dzięki krotkom funkcje w pythonie mogą zwracać kilka obiektów na raz.

````python
def kwadratowe(a, b, c):
    delta = b**2 - 4*a*c
    
    x1 = (-b + delta**(1/2))/(2*a)
    x2 = (-b - delta**(1/2))/(2*a)
    
    return (x1, x2) # nawiasy można też pominąć

print(kwadratowe(4, 1, 0))
````

### Rozpakowywanie
Elementy krotki czy listy można przypisać do kilku zmiennych, o ile ich liczba jest taka sama, jak liczba jej elementów.

````python
a = (1, 3, 2)
x, y, z = a
print(x)
print(y)
print(z)
````

Przydaje się to bardzo przy funkcjach zwracających krotki.

````python
x, y = kwadratowe(1, 2, 3)
print(x)
print(y)
````

# Lista `sys.argv`
Biblioteka `sys` posiada specjalną listę `sys.argv`. Gdy uruchamiamy skrypt pythonowy z wiersza poleceń, po jego nazwie możemy dopisać dodatkowe argumenty. `sys.argv` pozwala się do nich odwołać.

Na przykład gdy wykonamy w teminalu polecenie:
```
>python3 skrypt.py 123 abcd plik.txt
```
to w programie `skrypt.py` lista `sys.argv` będzie miała wartość:
```
["skrypt.py", "123", "abcd", "plik.txt"]
```

`sys.argv` zwykle wykorzystuje się, by podać programowi nazwy plików, na których ma pracować.
`sys.argv` jest najbardziej przydatne w skryptach `.py`, w notatnikach `.ipynb` raczej nie ma zastosowania.

# Obiekt `None`

Czasem przydaje się coś podobnego do zera, `''`, pustej listy itp., ale nie będącego żadnym z nich. Do tego celu stworzono obiekt `None`.

`None` często stosuje się jako standardową wartość opcjonalnego argumentu mającą zasygnalizować, że argument nie nie został użyty.

Sprawdzać, czy jakaś zmienna jest `None` najlepiej używając `is` albo `is not`:

````python
def wypisz(n, m=None):
    if m is None:
        for k in range(n):
            print(k)
    else:
        for k in range(n, m):
            print(k)
            
wypisz(3)
````

````python
wypisz(2, 6)
````

Dodatkowo funkcje nie posiadające wyrazu `return` zwracają `None`.

````python
def f(x):
    print(f'x={x}')
    
a = f(3)
print(a)
````

# Importowanie własnych skryptów jako bibliotek
Python pozwala importować własne skrypty jako biblioteki. Jeśli znajdują się w tym samym katalogu, co obecny program, po prostu da się je zaimportować.

````python
import Z_1_bobelkowe

print(Z_1_bobelkowe.sortowanie_bobelkowe([3, 2, -7, 14]))
````

Importowanie skryptu `.py` powoduje wykonanie całego kodu w środku (python nie wie, które fragmenty są potrzebne do zdefiniowania funkcji itp., więc robi wszystko).

Jeśli chcemy, by niektóre fragmenty kodu wykonywały się tylko, gdy jest on uruchamiany samodzielnie, a nie, gdy jest importowany, należy umieścić je wewnątrz specjalnego if-a:
```
if __name__ == "__main__":                                                
    # Ten kod wykona się tylko, gdy program jest uruchamiany samodzielnie.
    # Nie wykona się, gdy jest importowany.                               
```
Przydaje się to na przykład do oddzielenia części interaktywnej programu (`input`, `print` itp.) od definicji funkcji.
