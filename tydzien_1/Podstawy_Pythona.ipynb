# if/else

Czasem potrzebujemy, by program zachowywał się inaczej w zależności od jakiegoś warunku.
Do tego służy struktura if/else:
```
if warunek_1:                               
    # Kod, który się wykona, jeśli          
    # spełniony jest warunek_1              
elif warunek_2:                             
    # Kod, który się wykona, jeśli          
    # warunek_1 nie jest spełniony          
    # ale spełniony jest warunek_2          
elif warunek_3:                             
    # Kod, który się wykona, jeśli          
    # warunek_1 i warunek_2 nie są spełnione
    # ale spełniony jest warunek_3          
else:                                       
    # Kod, który się wykona, jeśli          
    # żadne warunki nie były spełnione      
```

Kod wewnąrz if/else musi być wcięty.

````python
n = float(input("Podaj liczbę:"))

if n > 0:
    print("n jest dodatnie.")
elif n < 0:
    print("n jest ujemne.")
else:
    print("n = 0")
````

Operacje porównywania:

|   |          |
|---|:--------:|
| > | większe  |
| < | mniejsze |
| >= | większe lub równe |
| <= | mniejsze lub równe |
| == | równe |
| != | różne |

# Funkcje

Funkcje pozwalają na wielokrotne wykorzystanie tego samego fragmentu kodu w różnych miejscach.
Przyjmują argumenty i zazwyczaj zwracają jakąś wartość.

Deklarowanie funkcji:
```
def nazwa_funkcji(argument1, argument2, argument3):
    # kod wynkonywany w funkcji                    
    return wynik                                   
```
Wywołanie funkcji:
```
x = nazwa_funkcji(a, b, c)
```

````python
def dodaj(a, b):
    c = a + b # zmienne a, b i c są widoczne tylko wewnątrz funkcji dodaj
    return c

print(dodaj(2, 3))
````

````python
# zmienne a, b i c są widoczne tylko wewnątrz funkcji dodaj
print(a)
````

Funkcje mogą wywoływać same siebie. Nazywamy to rekurencją.

````python
def pomnóż_przez_sumowanie(x, n): # w nazwach funkcji czy zmiennych da się stosować polskie znaki, ale nie jest to zalecane
    if n > 0:
        return x + pomnóż_przez_sumowanie(x, n-1) # x*n = x + x*(n-1)
    elif n == 0:
        return 0 # x*0 = 0
    elif n < 0:
        return - pomnóż_przez_sumowanie(x, -n) # x*(-n) = -x*n
    
print(pomnóż_przez_sumowanie(2, 3))
````

# Pętle
Pętle służą do wielokrotnego wykonywania raz za razem fragmentu kodu.

### Pętla while

Pętla wykonuje się, dopuki spełniony jest warunek. Składnia:
```
while warunek:                                                         
    # Kod, który wykonuje się raz za razem, póki spełniony jest warunek
```

````python
n = 10
while n >= 0:
    print(n)
    n -= 1
    
print("Koniec pętli.")
````

Pętlę while można przerwać komendą `break`.

````python
n = 10
while True:
    print(n)
    n -= 1
    if n < 0:
        break
````

### Pętla for po liczbach z przedziału

````python
for i in range(10): # range(n) oznacza liczby 0, 1, 2, ..., n-1
    print(i)
````

````python
for i in range(3, 10): # range(m, n) oznacza liczby m, m+1, m+2, ..., n-1
    print(i)
````

Oprócz `range`, pętlę for można wykonać też po elementach listy czy innych kontenerów, ale o tym będziemy mówić na następnych i późniejszych zajęciach.
