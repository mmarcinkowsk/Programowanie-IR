# Programowanie Obiektowe

Kiedy piszemy programy o jakichś obiektach (ciałach niebieskich, ludziach, pięciokątach foremnych itp.), przydaje się móc je jakoś reprezentować w naszym programie. Do tego celu służy programowanie obiektowe.

By móc tworzyć takie obiekty, definije się ich cechy za pomocą **klas**. Python posiada wiele wbudowanych klas, z któych już korzystaliśmy, na przykład kalsa `int` reprezentuje liczby całkowite, a klasa `set` zbiory.

W klasie definiujemy:
- wewnętrzne zmienne opisujące obiekt - tzw. **pola**,
- wewnętrzne funkcje, które może wykonywać obiekt - tzw. **metody**.

Do pól i metod danego obiektu odowłuje się składnią `obiek.nazwa_pola` i `obiekt.nazwa_metody(argumenty)`. Przykładowo zmienna tekstowa (string) posiada metodę `endswith`, która sprawdza, czy tekst kończy się jakimś napisem.

Wsztstkie metody **muszą** jako pierwszy argument przyjmować `self`. `self` reprezentuje obiekt, który definiujemy kalsą. Wewnątrz metod klasy odołujemy się do pól składnią `self.nazwa_pola`, a inne metody wywołujemy składnią `self.nazwa_metody(argumenty)`.
Dla osób znających C++, `self` jest analogiem do `this`.

Klasy powinny zawierać specjalną (magiczną) metodę `__init__` - tzw. konstruktor. Konstruktor opisuje, na czym polega tworzenie nowego obiektu danej klasy.

Składnia definiowania kalsy:

```python
class Osoba(): # zwyczajowo nazwy własnych klas powinny zaczynać się z wielkiej litery
    
    def __init__(self, imie, nazwisko, wiek): # konstruktor, przy tworzeniu nowego obiektu przyjmuje 3 argumenty: imie, nazwisko, wiek
        self.imie = imie # tworzę pole `imie`. `self` reprezentuje tworzony obiekt
        self.nazwisko = nazwisko
        self.wiek = wiek
        
    def czy_pelnoletnia(self): # metoda nie przyjmująca argumentów
        if self.wiek >= 18:
            return True
        else:
            return False
        
    def zmien_wiek(self, nowy_wiek): # metoda przyjmująca jeden argument
        self.wiek = nowy_wiek
```

Gdy zdefiniujemy klasę, jej obiekty tworzy się składnią `obiekt = NazwaKlasy(argumenty, dla, konstruktora)`.

```python
o1 = Osoba('Jan', 'Kowalski', 35)
o2 = Osoba('Anna', 'Kowalska', 38)
```

Do pól obiektu odwołuje się składnią `obiekt.nazwa_pola`.

```python
print(o1.imie)
```

Do pól można przypisywać nowe wartości, jak do normalnych zmiennych. Obiekt to zapamiętuje.

```python
o1.imie = 'Piotr'
print(o1.imie)
```

Do metod odołuje się składnią `obiekt.nazwa_metody()`.

```python
print(o1.czy_pelnoletnia())
```

Jeśli dokonamy zmian w obiekcie, metody to poczują.

```python
o1.zmien_wiek(15)
print(o1.czy_pelnoletnia())
```

# metody/pola publiczne i prywatne

Dotychczas stworzone metody i pola były publiczne - można się do nich odwołać spoza kalsy. Czasem jednak potrzebne jest stworzenie metod/pól wewnętrznych, które nie są potrzebne użytkownikowi klasy, a czasem nieostrożne zmienianie ich przez użytkownika może uszkodzić obiekt i zawiesić program.

Jeśli zaczniemy nazwę metody/pola od dwóch podkreśleń `__` i nie zakończymy jej dwoma podkreśleniami, Python uzna, że ta metoda/pole jest prywatna. Wtedy ciężej się do niej dostać spoza klasy, a nadal jest dostępna wewnątrz.

Do zmieniania wartości pól prywanych zazwyczaj tworzy się metody `set_nazwa`, które oprócz jej zmiany wykonują dodatkowe czynności, jakie powinny towarzyszyć tej zmianie. Do zwracania takiego pola zwykle tworzy się metody `get_nazwa`.

```python
class Osoba_z_peselem():
    
    def __init__(self, imie, nazwisko, wiek, pesel):
        self.imie = imie
        self.nazwisko = nazwisko
        self.wiek = wiek
        self.__pesel = pesel # pole prywantne
        
    def czy_pelnoletnia(self):
        if self.wiek >= 18:
            return True
        else:
            return False
        
    def zmien_wiek(self, nowy_wiek):
        self.wiek = nowy_wiek
        
    def get_pesel(self):
        return self.__pesel # pole prywatne jest dostępne wewnątrz klasy
    
o3 = Osoba_z_peselem('Adam', 'Nowak', 28, '12345678901')

print(o3.__pesel) # Python utrudnia odwoływanie się do pól prywatnych poza klasą
```

```python
print(o3.get_pesel()) # publiczna metoda może odczytać prywatne pole i zwrócić jego wartość
```

# Funkcje z dowolną liczbą argumentów

Przy definiowaniu funkcji możemy dodać na końcu specjlany argument `*args` (nazwa może być inna, ale ważna jest gwiazdka). Wtedy funkcja może przyjąć dowolną liczbę argumentów, a wszystkie "nadmiarowe" zostaną zebrane w krotkę, która będzie dostępna pod nazwą `args`.

```python
def f(x, y, *args):
    print(f'x={x}')
    print(f'y={y}')
    print(f'args={args}')
    
f(1, 2, 3, 4, 5, 6)
```
