# `enumerate`

`enumerate` przyjmuje listę (czy inny kontener, po którym można zrobić pętlę `for`) i zwraca pary (krotki) `(numer_elementu, element)`.

```python
lista = ['abc', 'efg', 'hij', 'klm']

for numer, napis in enumerate(lista):
    print(numer, napis)
```

Czasem wygodniej jest zrobić pętlę po `enumerate`, niż po `range(len(lista))`. Oprócz tego, gdy robimy pętlę po zawartości słownika, nie ma naturalnej kolejności, więc `enumerate` jest naturalnym sposobem by wiedzieć, w którym powtórzeniu pętli jesteśmy.

Można też stworzyć zmienną całkowitą, której wartość zwiększamy przy każdym powtórzeniu. Jednak wtedy możemy popełnić błąd i ją zmodyfikować. `enumerate` oblicza wartość tej liczby przy każdym powtórzeniu, więc taki błąd by się nie propagował.

# Wyrażenia regularne (regex)

Wyrażenia regularne (regex) są narzędziem do znajdowania fragmentów napisów pasujących do szukanego przez nas wzorca. Pozwalają też na sprecyzowanie, które fragmeny napisu nas interesują i automatyczne wydobycie ich z napisu. Regex jest dostępny w większości języków programowania oraz w wielu programach (np. `grep`). Jest też dostępne wiele stron internetowych, które pozwalają na projektowanie i testowanie regexów, na przykład https://regex101.com/.

Wzory presyzuje się napisami, składającymi się z liter czy cyfr, do których napis jest dopasowywany bezpośrednio, oraz znaków specjalnych. Najpopularniesze znaki specjalne:

- `.` - dowolny symbol,
- `A+` - `A` powtórzone przynajmniej raz,
- `A*` - `A` powtórzone dowolną liczbę razy (w tym zero),
- `A?` - `A` raz lub zero razy,
- `[abcd]` - dowolny znak będący `a`, `b`, `c` lub `d`,
- `[^abcd]` - dowolny znak nie będący `a`, `b`, `c` lub `d`,
- `[a-g]` - dowolna litera od `a` do `g`,
- `A|B` - `A` lub `B`,
- `\.`, `\*` itd. - znak `.`, `*` itd. rozumiany dosłownie, a nie w specjalnym znaczeniu jak powyżej,
- `\d` - cyfra, tożsame z `[0-9]`,
- `(coś)` - zaznaczenie grupy, która nas interesuje i ma zostać wydobyta.

W Pythonie narzędzie regex implementuje wbydowana biblioteka `re`.

`re.search(wzór, napis)` szuka fragmentu stringa `napis` spełniającego wzór `wzór`. Jeśli go nie znajdzie, zwraca `None`. Jeśli go znajdzie, zwraca specjalny obiekt zawierający informacje o pasującym fragmencie napisu i grupach.

Poniważ `None` ma wartość logiczną fałsz, obiekt dopasowania zaprojektowano tak, że zawsze ma wartość logiczną prawda. Dzięki temu łatwo sprawdzić, czy regex dopasował wzór.

```python
import re

wzor = r'[a-z]+'

napis = '123python456'

dopasowanie = re.search(wzor, napis)

if dopasowanie:
    print(dopasowanie)
    print(dopasowanie.group(0)) # grupa 0 to całe dopasowanie
```

Przykład regexa, który znajduje liczbę rzeczywistą zapisaną w formacie naukowym i w grupach zwraca jej mantysę i wykładnik:

(być może minus; przynajmniej jedna cyfra; być może kropka; dowolna liczba cyfr) e lub E (być może minus; przynajmniej jedna cyfra)

```python
wzor = r'(-?\d+\.?\d*)[eE](-?\d+)'

dopasowanie = re.search(wzor, '1.256e-3')

if dopasowanie:
    print(dopasowanie)
    print(dopasowanie.group(1))
    print(dopasowanie.group(2))
```

Regexy przydają się na przykład przy wczytywaniu plików tekstowych, w których w nietypowy sposób zapisane są dane.
