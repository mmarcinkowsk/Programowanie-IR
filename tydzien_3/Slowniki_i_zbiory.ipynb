# Słowniki i zbiory
Słowniki i zbiory to inne rodzaje kontenerów w Pythonie. Od list i krotek różni je przede wszystkim to, że kolejność elementów nie jest ważna.

### Słowniki
Słowniki przechowują pary `klucz:wartość` i pozwalają się odwołać do konkretnej wartości za pomocą jej klucza. Klucze mogą być stringami, liczbami i wieloma innymi obiektami.
Słowniki tworzy się, zamykając pary `klucz:wartość` w nawiasy klamrowe.

```python
slownik = {"Ala ma kota":1, 2:3, "abc":"cdf", None:print, 4:[1, 2, 3]}
print(slownik["Ala ma kota"])
```

```python
print(slownik["abc"])
```

```python
print(slownik[4])
```

Wartość przy danym kluczu można zmienić. Jednocześnie oznacza to, że każdy klucz występuje tylko raz.

```python
slownik["abc"] = "ghi"
print(slownik["abc"])
```

Nowy klucz można dodać, po prostu odwołując się do niego i przypisując wartość.

```python
slownik["Nowy klucz którego jeszcze nie było"] = "Związana z nim wartość"
print(slownik["Nowy klucz którego jeszcze nie było"])
```

Jeżeli spróbujemy odczytać wartość z nieistniejącego klucza, Python wyrzuci błąd.

```python
print(slownik["Tego klucza nie ma"])
```

Słowniki mają za to metodę `.get(klucz)`, która zwraca wartość odpowiadającą istniejącemu kluczowi, a przy nieistenijącym zwaraca `None`.

```python
print(slownik.get("abc"))
print(slownik.get("Tego klucza nie ma"))
```

Ze słownika można usunąć element, korzystając z metody `.pop(klucz)`.

Do słownika można dołączyć inny słownik korzystając z metody `.update()`.

```python
slownik.update({7:"siedem", "i":"j"})
print(slownik)
```

### Pętle po słownikach
Do robienia pętli po słownikach służą 3 metody.

`slownik.keys()` służy do robienia pętli po kluczach.

```python
for klucz in slownik.keys():
    print(klucz)
```

`slownik.values()` służy do robienia pętli po wartościach.

```python
for wartosc in slownik.values():
    print(wartosc)
```

`slownik.items()` pozwala na robienie pętli po krotkach `(klucz, wartość)`.

```python
for klucz, wartosc in slownik.items():
    print(f"Klucz:{klucz}, wartość:{wartosc}")
```

### Zagnieżdżone słowniki
Bardzo często w słownikach umieszcza się inne słowniki. Wtedy, ponieważ `slownik[klucz]` jest słownikiem, to można się odwołać do niego odwołać, jak do zwykłedo słownika:

`slownik[klucz][wewenętrzny_klucz]`

```python
uczelnie = {"Uniwersytet Warszawski":{"miasto":"Warszawa", "ulica":"Krakowskie Przedmieście 26/28", "kod pocztowy":"00-927"}, "Politechnika Warszawska":{"miasto":"Warszawa", "ulica":"Politechniki 1", "kod pocztowy":"00-661"}}
print(uczelnie["Uniwersytet Warszawski"]["ulica"])
```

### Wyrażenia słownikowe
Podobnie jak wyrażenia listowe, słowniki można tworzyć pętlami `for` "zamniętymi w słowniku".

```python
print({x:x**2 for x in range(10)})
```

### Zbiory
Zbiory w pythonie są analogią do zbiorów matematycznych. Coś albo należy do zbioru, albo nie.

Zbiory **nie zapamiątują kolejności** elementów. Nie zapamiątują też, ile razy dany element był do nich włożony.

Zbiory tworzy się, zamykając elementy w nawiasy klamrowe.

```python
zbior = {1, 5, 8, 8, 8, "Ala ma kota", 19, -0.3, None}
print(zbior)
```

Przynależność obiektu do zbioru sprawdza się słowem `in`.

```python
print(1 in zbior)
print(12 in zbior)
```

Do zbiorów można dodawać nowe elementy metodą `.add()` i usuwać metodą `.remove()`.

```python
zbior.add(2703)
zbior.remove(5)
print(zbior)
```

Po elementach zbioru można zrobić pętlę. Kolejność, w jakiej elementy będą brane ze zbioru **nie zależy** od kolejności, w jakiej je do zbioru dodano.

```python
for element in zbior:
    print(element)
```

Jak listy czy słowniki, zbiory można tworzyć wyrażeniami zbiorowymi.

```python
print({n for n in range(-5, 5)})
```

Na zbiorach można wykonywać operacje sumy mnogościowej $A\cup B$, przecięcia $A\cap B$, różnicy $A\backslash B$ i różnicy symetrycznej $A\Delta B=(A\backslash B)\cup(B\backslash A)$.

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
print(A|B)
print(A&B)
print(A-B)
print(A^B)
```

Można też sprawdzać, czy jakiś zbiór jest podzbiorem innego zbioru.

```python
C = {4, 5}
print(C < A)
print(C < B)
print(C < C)
print(C <= C)
```

# Funkcje `list()`, `tuple()`, `set()` i `dict()`

Dotychczas wykorzystywaliśmy funcje `int()` i `float()`, żeby przerabiać inne obiekty na liczby całkowite i rzeczywiste. Kontenery w Pythonie posiadają podobne funkcje.

```python
list((1, 2, 3))
```

```python
tuple([1, 2, 3])
```

```python
set([1, 2, 3, 4, 2])
```

```python
dict([('a', 1), ('b', 2)])
```

Jeśli wywołamy je bez argumentów, funkcje te tworzą puste kontenery.

```python
print(list())
print(tuple())
print(set())
print(dict())
```

# Nie wszystko można włożyć do zbioru (czy uczynić kluczem słownika)

Niektórych typów obiektów nie można umieszczać w zbiorach. W szczególności nie można tam wkładać obiektów, które w trakcie życia programu mogą ulegać zmianie, jak listy, słowniki czy inne zbiory.

```python
{1, [1, 2, 3], "b"}
```

Tych samych typów nie można używać jako kluczy w słownikach.

```python
{"a":"b", {1, 2, 3}:"d", 3:4}
```

Krotki są niezmiennicze, więc mogą służyć do obu celów.

```python
print({1, (1, 2, 3), "b"})
print({"a":"b", (1, 2, 3):"d", 3:4})
```
